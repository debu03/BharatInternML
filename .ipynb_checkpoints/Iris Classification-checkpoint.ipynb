## Iris Classification

Iris classification is a typical machine learning classifictaion project where we use a dataset to train the module to identify/recognise the target. There are three species of iris flower. When a new flower is given, we need to predict it belongs to which type. From the sample of three species ie. `Sentosa`, `Versicolour`, `Virginica`. We don't have to use image processing.

### Import Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import sklearn
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

### Loading Dataset

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
# Removing unecessary data ie. Id...
df=df.drop(['Id'],axis=1)
df.head()
```

### Model Traning and Testing

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import pickle
```

```python
X=df.drop(['Species'],axis=1)
y=df['Species']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
# LogisticRegression model...(from map)
model= LogisticRegression()
model.fit(X_train.values,y_train.values)
model.score(X_test.values,y_test.values)*100
```

```python
# K-nearest Neighbors model to improve accuracy...
model= KNeighborsClassifier()
model.fit(X_train.values,y_train.values)
model.score(X_test.values,y_test.values)*100
```

```python
# Decision Tree model...

model= DecisionTreeClassifier()
model.fit(X_train.values,y_train.values)
model.score(X_test.values,y_test.values)*100
```

### Save Model

```python
import pickle
```

```python
filename='iris_model.sav'
try:
    with open(filename,'wb') as file:
        pickle.dump(model,file)
    print('Model saved successfully')
except Exception as e:
    print(f'Error saving model: {e}')
    
```

```python
load_model=pickle.load(open(filename,'rb'))
```

```python
load_model.predict([[6.0,2.2,4.0,1.0]])
```
